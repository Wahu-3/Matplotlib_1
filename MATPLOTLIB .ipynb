```python
import pandas as pd
```

```python
df = pd.read_csv("jumta.csv")
```

```python
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
line_plot_revenue = df['Revenue'].values  
plt.plot(line_plot_revenue)
plt.xlabel('Month') 
plt.ylabel('Revenue')
plt.title('Line Plot of Revenue')
plt.grid()
plt.show()
```

```python
#bar plot showing the total number of sales
plt.figure(figsize=(20,15))
orders = df["Orders"]
plt.bar(df["Month"], df["Orders"])
plt.title('Bar Plot of Total Orders')
plt.xlabel("Month")
plt.ylabel('Orders')
plt.xticks(rotation=45)
plt.show()
```

```python
#scatter plot
import matplotlib.pyplot as plt
average_order = df['Orders'].mean()
plt.scatter(df["Orders"], [average_order] * len(df), label='Average Orders') 
plt.xlabel('Orders')
plt.ylabel('Average Orders')
plt.title('Scatter Plot of Orders vs Average Orders')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
# Plot 1: Line Plot of Revenue
plt.subplot(3, 1, 1)
line_plot_revenue = df['Revenue'].values
plt.plot(line_plot_revenue)
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.title('Line Plot of Revenue')
plt.grid()

# Plot 2: Bar Plot of Total Orders
plt.figure(figsize=(20,15))
plt.subplot(3, 1, 2)
orders = df["Orders"]
plt.bar(df["Month"], df["Orders"])
plt.xlabel("Month")
plt.ylabel('Orders')
plt.title('Bar Plot of Total Orders')
plt.xticks(rotation=45)

# Plot 3: Scatter Plot of Orders vs Average Orders
plt.subplot(3, 1, 3)
average_order = df['Orders'].mean()
plt.scatter(df["Orders"], [average_order] * len(df), label='Average Orders')
plt.xlabel('Orders')
plt.ylabel('Average Orders')
plt.title('Scatter Plot of Orders vs Average Orders')
plt.legend()

# Show the plot
plt.show()
```
